```python
file = pd.read_csv("Mapped_set_final(Imputed).csv")
coltorep=['vg0130976864', 'vg0135617816', 'vg0138428840', 'vg0405463422',
       'vg0713178880', 'vg0819793460', 'vg1019044175', 'vg1123563633',
       'vg1207667840']
file = file.set_index("Cultivar ID")
for col in coltorep:
    primary_val = file.loc['Primary', col]
    file[col] = file[col].replace("N", primary_val)
for col in coltorep:
    secondary_val = file.loc['Secondary', col]
    file[col] = file[col].replace("DEL", secondary_val)
file = file.iloc[6:, :]
file.reset_index(inplace = True)
for i in range(len(file['Subpopulation'])):
    if file['Subpopulation'][i] == "Indica II":
        m = file['Plant Height (cm)'][i].mean()
file['Plant Height (cm)'] = file['Plant Height (cm)'].replace(to_replace = -9.0, value = m)
file.to_csv("Updated.csv", index = True)
```
