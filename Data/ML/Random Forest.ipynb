```python
import pandas as pd
import numpy as np
file = pd.read_csv("Database.csv")
file
```

```python
legend = {'A': 1,'T': 2,'C':3,'G': 4}
encoded=[]
for i in file['Gene Sequence']:
  score = ""
  for j in i:
    score += str(legend[j])
  encoded.append(int(score))
```

```python
del file['Unnamed: 0']
```

```python
file['Gene Sequence'] = encoded
```

```python
file
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# prepare data for training
X = file['Gene Sequence'].values
y_subpop = file['Subpopulation'].values

# split the data into training and test sets
X_train, X_test, y_train_subpop, y_test_subpop = train_test_split(X, y_subpop, test_size = 0.28, random_state = 42)
X_train = X_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)

# fit the random forest classifier for predicting subpopulation
rfc_subpop = RandomForestClassifier(n_estimators = 100, random_state = 42, max_depth = 10)
rfc_subpop.fit(X_train, y_train_subpop)

# predict the test set for subpopulation
y_pred_subpop = rfc_subpop.predict(X_test)

# calculate accuracy score for subpopulation prediction
acc_score_subpop = accuracy_score(y_test_subpop, y_pred_subpop)
print("Accuracy score for subpopulation prediction: {:.0f}%".format(acc_score_subpop*100))
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# extract features and target variable
x = file['Gene Sequence'].values
y = file['Plant Height(cm)'].values

# split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 42)
x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)
# fit the regressor
regressor = RandomForestRegressor(n_estimators = 100, random_state = 42, max_depth = 10)
regressor.fit(x_train, y_train)

# predict on the test set
y_pred = regressor.predict(x_test)

# evaluate the model
from sklearn.metrics import r2_score
print("R^2 score: {:.0f}".format(r2_score(y_test, y_pred)*100))
```
