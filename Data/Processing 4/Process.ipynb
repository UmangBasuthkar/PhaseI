```python
del file['Unnamed: 0']
```

```python
file = file.rename(columns = {"index" : "Clustar Number", "cultivar id": "Cultivar ID"
                              ,"sequences" : "Gene Sequence", "height" : "Plant Height(cm)"
                              ,"subpopulation" : "Subpopulation"})
```

```python
file
```

```python
import numpy as np
height_range = []
for i in range(68):
    final_height = 0
    plant_height = []
    for j in range(len(file['Clustar Number'])):
        if (file['Clustar Number'][j] == i):
            plant_height.append(file['Plant Height(cm)'][j])
    final_height = np.mean(plant_height)
    for k in range(len(plant_height)):
        height_range.append(final_height)
```

```python
from statistics import mode
common_sub = []
for i in range(68):
    sub_name = []
    for j in range(len(file['Clustar Number'])):
        if (file['Clustar Number'][j] == i):
            sub_name.append(file['Subpopulation'][j])
    if (len(sub_name) == 2):sub_name[1] = sub_name[0]
    else:
        sub_mode = mode(sub_name)
        for k in range(len(sub_name)):sub_name[k] = sub_mode
    for k in range(len(sub_name)):
        common_sub.append(sub_name)
```

```python
database = pd.DataFrame(columns = ['Gene Sequence', 'Cultivar ID', 'Plant Height(cm)', 'Subpopulation'])
for i in range(len(file['Gene Sequence'])):
    database.loc[i, 'Gene Sequence'] = file.loc[i, 'Gene Sequence']
    database.loc[i, 'Cultivar ID'] = file.loc[i, 'Cultivar ID']
    database.loc[i, 'Plant Height(cm)'] = height_range[i]
    database.loc[i, 'Subpopulation'] = common_sub[i][0]
database.to_csv("Database.csv")
```
