```python
from Bio import AlignIO
align = AlignIO.read("output.aln", "clustal")
clusters = {}
for record in align:
    seq = record.seq
    if seq not in clusters:
        clusters[seq] = []
    clusters[seq].append(record.id)
```

```python
import pandas as pd
data = pd.DataFrame(columns = ['Sequences','Cultivar ID', 'Subpopulation', 'Plant Height(cm)'])
for records in align:
    values = records.id.split("|")
    row = {"Sequences" : str(records.seq), "Cultivar ID" : values[1], "Subpopulation" : values[2], "Plant Height(cm)" : float(values[0])}
    data = pd.concat([data, pd.DataFrame([row])], ignore_index = True)
data.head(10)
```

```python
with open("Clusters.csv", "w") as file2:
    for key, value in clusters.items():
        a = ""
        a = (key, value)
        a = str(a)
        file2.write(a + "\n")
```
